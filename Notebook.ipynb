```python
data.columns
```

```python
x = data[['number of bedrooms', 'number of bathrooms',
       'living area', 'condition of the house', 'Number of schools nearby']]
y = data[["Price"]]       
```

```python
x
```

```python
y
```

```python
data.shape[0] * 0.8
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.2)
```

```python
x_train.shape
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import GridSearchCV
```

```python
from sklearn.tree import DecisionTreeRegressor
rfr = RandomForestRegressor()
param_grid = {
    "criterion" : ["mse", "friedman_mse", "mse"],
    "splitter" : ["best", "random"],
    "max_depth" : [None, 10, 20, 30, 40, 50],
    "min_samples_split" : [2,5,10],
    "min_samples_leaf" : [1,2,4]
    }
```

```python
gridrfr = GridSearchCV(
    estimator=rfr,
    param_grid=param_grid,
    cv=5,
    scoring="r2",
    n_jobs=-1
)
```

```python
tree_model = DecisionTreeRegressor()
```

```python
grid_tree = GridSearchCV(estimator = tree_model,param_grid = param_grid)
```

```python
grid_tree.fit(x_train, y_train)
```

```python
grid_tree.best_params_
```

```python
tree_preds = grid_tree.predict(x_test)
```

```python
from sklearn.metrics import mean_squared_error
mean_squared_error(y_test, tree_preds)
```

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
```

```python
lr.fit(x_train, y_train)
```

```python
predslr = lr.predict(x_test)
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
```

```python
mean_absolute_error(y_test, predslr)
```

```python
from sklearn.ensemble import RandomForestRegressor
rfrmodel = RandomForestRegressor()
```
